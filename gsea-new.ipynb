```python
from sanbomics.tools import id_map
```

```python
import pandas as pd
from gseapy.plot import gseaplot
import gseapy as gp
import numpy as np
from pydeseq2.dds import DeseqDataSet
from pydeseq2.ds import DeseqStats
import pandas as pd
```

```python
counts = pd.read_csv('Merged.csv')
counts
```

```python
counts = counts.dropna()
```

```python
counts = counts.set_index('ENSG_ID')
counts
```

```python
counts = counts.T
```

```python
metadata = pd.read_csv("annotation.csv")
metadata = metadata.set_index("Samples")
metadata
```

```python
dds = DeseqDataSet(counts=counts,
            metadata=metadata,
            design_factors="Group")
```

```python
dds.deseq2()
```

```python
stat_res = DeseqStats(dds,contrast = ('Group','Normal','Diseased'))
```

```python
stat_res.summary()
```

```python
res = stat_res.results_df
```

```python
res
```

```python
mapper = id_map(species = 'human')
```

```python
res['Symbol'] = res.index.map(mapper.mapper) 
```

```python
res
```

```python
df = res
```

```python
df = df.rename(columns = {'Symbol' : 'Gene'})
```

```python
df
```

```python
import numpy as np
```

```python
df['Rank'] = -np.log10(df.padj)*df.log2FoldChange
```

```python
df = df.sort_values('Rank', ascending = False).reset_index(drop = True)
```

```python
df
```

```python
ranking = df[['Gene', 'Rank']]
```

```python
ranking
```

```python
ranking = ranking.dropna()
```

```python
ranking
```

```python
gp.get_library_name()
```

```python
pre_res = gp.prerank(rnk = ranking, gene_sets = 'GO_Biological_Process_2021', seed = 6, permutation_num = 100)
```

```python
out = []

for term in list(pre_res.results):
    out.append([term,
               pre_res.results[term]['fdr'],
               pre_res.results[term]['es'],
               pre_res.results[term]['nes']])

out_df = pd.DataFrame(out, columns = ['Term','fdr', 'es', 'nes']).sort_values('fdr').reset_index(drop = True)
out_df
```
