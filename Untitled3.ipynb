```python
from pydeseq2.dds import DeseqDataSet
from pydeseq2.ds import DeseqStats

import pandas as pd
```

```python
counts = pd.read_csv('count_matrix_new.csv')
counts
```

```python
counts = counts.set_index('Geneid')
counts
```

```python
counts = counts[counts.sum(axis = 1) > 0]
counts
```

```python
counts = counts.T
```

```python

counts
```

```python
metadata = pd.DataFrame(zip(counts.index, ['P', 'P', 'P', 'P', 'P', 'P', 'C', 'C', 'P', 'P', 'P', 'P','C', 'C','C', 'C','C', 'C','P', 'P', 'P', 'P', 'C']),
                        columns = ['Sample', 'Condition'])
```

```python
metadata = metadata.set_index('Sample')
metadata
```

```python
dds = DeseqDataSet(counts=counts,
            metadata=metadata,
            design_factors="Condition")
```

```python
dds.deseq2()
```

```python
dds
```

```python
stat_res = DeseqStats(dds,contrast = ('Condition','P','C'))
```

```python
stat_res.summary()
```

```python
res = stat_res.results_df
```

```python

res
```
